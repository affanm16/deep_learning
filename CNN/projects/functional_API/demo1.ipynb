```python
from keras.models import Model
```

this is based on a simple problem where we have salary,height,marital status and we have to predict age and place i.e delhi/mumbai (classification) 3 inputs 2 outputs

```python
from keras.layers import *
```

```python
x=Input(shape=(3,))
```

you must give a name to the layer (here hidden1) and specify from where you are getting the input/it is connected which layer(here x)

```python
hidden1=Dense(128,activation='relu')(x)
hidden2=Dense(64,activation='relu')(hidden1)

#branching
output1=Dense(1,activation='linear',name='age')(hidden2)
output2=Dense(1,activation='sigmoid',name='place')(hidden2)
```

```python
model=Model(inputs=x,outputs=[output1,output2])
```

```python
model.summary()
```

```python
#visualizing
from keras.utils import plot_model
```

```python
plot_model(model)
```

```python
plot_model(model,show_shapes=True)
```
